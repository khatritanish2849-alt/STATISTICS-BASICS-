**Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**

Answer — Python code and output:

   - # Q5: mean, median, mode
    import statistics as stats

    numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

    mean_numbers = stats.mean(numbers)
    median_numbers = stats.median(numbers)
# multimode handles multiple modes
    modes_numbers = stats.multimode(numbers)

    print("Mean:", mean_numbers)
    print("Median:", median_numbers)
    print("Mode(s):", modes_numbers)

Expected output (computed):

    Mean: 19.6
    Median: 19
    Mode(s): [12, 19, 24]

 Explanation: There are three modes (12, 19, 24) each appearing three times. Mean is the arithmetic average, and median is the middle value of the sorted list.



**Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]**

Answer — Python code and output:

# Q6: covariance and correlation
    import numpy as np

    list_x = [10, 20, 30, 40, 50]
    list_y = [15, 25, 35, 45, 60]

# Sample covariance (ddof=1)
    cov_xy = np.cov(list_x, list_y, ddof=1)[0, 1]
# Pearson correlation coefficient
    corr_xy = np.corrcoef(list_x, list_y)[0, 1]

    print("Covariance (sample):", cov_xy)
    print("Correlation coefficient (Pearson r):", corr_xy)

Expected output (computed):

    Covariance (sample): 275.0
    Correlation coefficient (Pearson r): 0.995893206467704

Interpretation: The covariance is positive, indicating that as x increases, y tends to increase. The Pearson correlation is about 0.996, very close to 1, indicating a very strong positive linear relationship.


**Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]**

Answer — Python code and explanation:

# Q7: boxplot and outlier detection using IQR
    import numpy as np
    import matplotlib.pyplot as plt

    data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create boxplot (this will display the boxplot when run locally)
    plt.boxplot(data, vert=False)
    plt.title("Boxplot of data")
    plt.xlabel("Value")
    plt.show()

# Compute IQR and fences
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr

    outliers = [x for x in data if x < lower_fence or x > upper_fence]

    print("Q1:", q1)
    print("Q3:", q3)
    print("IQR:", iqr)
    print("Lower fence:", lower_fence)
    print("Upper fence:", upper_fence)
    print("Outliers:", outliers)

Expected numeric results (computed):

    Q1: 17.25
    Q3: 23.25
    IQR: 6.0
    Lower fence: 8.25
    Upper fence: 32.25
    Outliers: [35]

  Explanation: Using the IQR method (1.5 × IQR), the upper fence is 32.25. The only value greater than this is 35, so 35 is an outlier. The rest of values lie within the typical range. The boxplot will show the box from Q1 to Q3 with the median inside, whiskers to min/max within fences, and a point for the outlier at 35.




**Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
Explain how you would use covariance and correlation to explore this relationship.
Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]**

#Answer — Explanation:

Covariance measures the joint variability of two variables. A positive covariance indicates that when advertising spend increases, daily sales tend to increase; a negative covariance indicates the opposite. However, covariance is in units that depend on the variables, so it is hard to interpret its magnitude directly.

Correlation (Pearson r) standardizes covariance by the product of the standard deviations of the two variables, giving a unitless measure between -1 and 1. A correlation near +1 indicates a strong positive linear relationship; near -1 indicates a strong negative linear relationship; near 0 indicates little or no linear relationship. Use correlation to quantify the strength and direction; use covariance to check direction and compute linear relationships (e.g., in regression).

Additionally, visualize with a scatter plot (advertising spend on x-axis, daily sales on y-axis). Consider computing a linear regression line and checking residuals to examine whether the relationship is approximately linear.

#Answer — Python code and output:

# Q8: covariance and correlation between advertising spend and daily sales
import numpy as np

    advertising_spend = [200, 250, 300, 400, 500]
    daily_sales = [2200, 2450, 2750, 3200, 4000]

    cov_ad_sales = np.cov(advertising_spend, daily_sales, ddof=1)[0, 1]
    corr_ad_sales = np.corrcoef(advertising_spend, daily_sales)[0, 1]

    print("Covariance (sample):", cov_ad_sales)
    print("Correlation coefficient (Pearson r):", corr_ad_sales)

#Expected output (computed):

    Covariance (sample): 84875.0
    Correlation coefficient (Pearson r): 0.9935824101653329

  Interpretation: Correlation ≈ 0.9936 indicates a very strong positive linear relationship between advertising spend and daily sales in this sample. This suggests increases in advertising are associated with increases in sales. (Note: correlation ≠ causation; consider experimental or controlled analyses to establish causality, and check for confounders.)


**Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.
Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**

#Answer — Explanation:

Use the following summary statistics and visuals:
Summary statistics:

1.Mean and median (central tendency).

2.Mode (most frequent score).

3.Standard deviation and variance (spread).

4.Minimum and maximum (range).

5.Percentiles / IQR (25th, 75th percentiles and IQR) to understand skewness and spread.

Visualizations:

1.Histogram to view the distribution (frequency by score or bin).

2.Boxplot to spot outliers and visualize quartiles.

3.Bar chart of counts per each score (1–10) if scores are integers.

4.Optionally, a density plot for a smooth estimate (if many observations).

#Answer — Python code and output:

# Q9: summary stats and histogram
import statistics as stats

import matplotlib.pyplot as plt

    survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

    mean_s = stats.mean(survey_scores)
    median_s = stats.median(survey_scores)
    modes_s = stats.multimode(survey_scores)
    stdev_s = stats.stdev(survey_scores)  # sample standard deviation
    variance_s = stats.variance(survey_scores)
    min_s = min(survey_scores)
    max_s = max(survey_scores)

    print("Mean:", mean_s)
    print("Median:", median_s)
    print("Mode(s):", modes_s)
    print("Standard deviation (sample):", stdev_s)
    print("Variance (sample):", variance_s)
    print("Min:", min_s, "Max:", max_s)

# Histogram (this will show a histogram when run locally)

    plt.hist(survey_scores, bins=range(3,12), align='left', rwidth=0.8)
    plt.xticks(range(4,11))
    plt.xlabel("Survey score")
    plt.ylabel("Frequency")
    plt.title("Histogram of survey scores")
    plt.show()

#Expected numeric results (computed):

    Mean: 7.333333333333333
    Median: 7
    Mode(s): [7]
    Standard deviation (sample): 1.632993161855452
    Variance (sample): 2.6666666666666665
    Min: 4 Max: 10

  Interpretation: The mean is about 7.33 and median is 7, indicating central tendency around 7. The mode is 7 (most common). Standard deviation ≈ 1.633 suggests moderate spread around the mean. The histogram will show which scores are most frequent — here scores 7, 8, and 9 appear frequent. There are no extreme outliers (range 4–10), but visualizing bin counts will confirm the shape (likely slightly right-tailed or roughly symmetric around 7).
